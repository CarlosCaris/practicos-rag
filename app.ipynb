```python
# load PDF
from langchain_community.document_loaders import PyPDFLoader
from dotenv import load_dotenv

load_dotenv()

file_path = (
    "data/D.S-N----977actualizado-2013.pdf"
)
loader = PyPDFLoader(file_path)
docs = loader.load()
docs[0:2]
```

```python
# split
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
splits = text_splitter.split_documents(docs)
splits[0:2]
```

```python
# data clean
import spacy
import re
from unidecode import unidecode

# Cargar el modelo de spaCy para el español
nlp = spacy.load("es_core_news_sm")

def clean_text(text):
    # 1. Conversión a minúsculas
    text = text.lower()
    
    # 2. Eliminación de caracteres especiales (excepto espacios y caracteres importantes)
    text = re.sub(r"[^a-záéíóúñü\s]", "", text)
    
    # 3. Normalización (eliminación de acentos)
    text = unidecode(text)
    
    # 4. Procesamiento NLP: Eliminación de stop words y lematización
    doc = nlp(text)
    clean_tokens = [
        token.lemma_ 
        for token in doc 
        if not token.is_stop and not token.is_punct
    ]
    
    # Unir los tokens limpios en una cadena
    return " ".join(clean_tokens)

# Aplicar la limpieza a los documentos divididos
for i, doc in enumerate(splits):
    splits[i].page_content = clean_text(doc.page_content)

# Verificar los resultados
splits[0:2]
```

```python
import os
import json
import numpy as np
from pathlib import Path
from transformers import AutoTokenizer, AutoModel
import torch

# Crear directorios
os.makedirs("data/chunks", exist_ok=True)
os.makedirs("embeddings", exist_ok=True)

# Cargar modelo de embeddings
model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Función para generar embeddings
def generate_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
        embeddings = outputs.last_hidden_state.mean(dim=1)  # Promedio de los tokens
    return embeddings[0].numpy()

# Segmentación y generación de embeddings
chunk_size = 1000  # Tamaño máximo en caracteres por chunk
chunk_overlap = 200  # Superposición entre chunks

chunk_data = []  # Para almacenar texto segmentado
embeddings = []  # Para almacenar embeddings

for i, doc in enumerate(splits):
    text = doc.page_content
    start = 0
    while start < len(text):
        # Crear fragmentos
        end = start + chunk_size
        chunk = text[start:end]
        
        # Ajustar hasta el último espacio antes de cortar
        if end < len(text):
            last_space = chunk.rfind(" ")
            if last_space != -1:
                chunk = chunk[:last_space]
                end = start + len(chunk)
        
        # Agregar fragmento al dataset
        chunk_data.append({
            "doc_id": i,
            "start": start,
            "end": end,
            "text": chunk.strip()
        })
        
        # Generar embedding del fragmento
        embedding = generate_embedding(chunk.strip())
        embeddings.append(embedding)
        
        # Avanzar con superposición
        start = end - chunk_overlap

# Guardar chunks en formato JSON
chunks_path = "data/chunks/chunks.json"
with open(chunks_path, "w", encoding="utf-8") as f:
    json.dump(chunk_data, f, ensure_ascii=False, indent=4)

# Guardar embeddings como numpy array
embeddings_path = "embeddings/embeddings.npy"
np.save(embeddings_path, np.array(embeddings))

print(f"Chunks guardados en: {chunks_path}")
print(f"Embeddings guardados en: {embeddings_path}")
```

```python
import os
from pathlib import Path
import json
import pickle
from datetime import datetime
from sentence_transformers import SentenceTransformer
import numpy as np
from tqdm import tqdm

class TextProcessor:
    def __init__(self, model_name='all-MiniLM-L6-v2'):
        self.embedding_model = SentenceTransformer(model_name)
        
    def semantic_chunk(self, splits, min_length=100):
        """
        Realiza una segmentación semántica basada en oraciones
        """
        # Usar spaCy para dividir en oraciones
        semantic_chunks = []
        
        for split in splits:
            doc = nlp(split.page_content)
            current_chunk = []
            current_length = 0
            
            for sent in doc.sents:
                sent_text = sent.text.strip()
                sent_length = len(sent_text.split())
                
                if current_length + sent_length > min_length:
                    # Guardar el chunk actual si tiene contenido
                    if current_chunk:
                        chunk_text = " ".join(current_chunk)
                        semantic_chunks.append({
                            "content": chunk_text,
                            "metadata": split.metadata
                        })
                    current_chunk = [sent_text]
                    current_length = sent_length
                else:
                    current_chunk.append(sent_text)
                    current_length += sent_length
            
            # Guardar el último chunk si tiene contenido
            if current_chunk:
                chunk_text = " ".join(current_chunk)
                semantic_chunks.append({
                    "content": chunk_text,
                    "metadata": split.metadata
                })
        
        return semantic_chunks

    def generate_embeddings(self, chunks):
        """
        Genera embeddings para cada chunk usando Sentence Transformers
        """
        texts = [chunk["content"] for chunk in chunks]
        embeddings = self.embedding_model.encode(texts, show_progress_bar=True)
        
        # Combinar chunks con sus embeddings
        for chunk, embedding in zip(chunks, embeddings):
            chunk["embedding"] = embedding.tolist()
        
        return chunks

    def save_processed_data(self, chunks, base_path="data"):
        """
        Guarda los chunks y embeddings en sus respectivas carpetas
        """
        # Crear directorios si no existen
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        chunks_dir = Path(base_path) / "chunks"
        embeddings_dir = Path(base_path) / "embeddings"
        
        chunks_dir.mkdir(parents=True, exist_ok=True)
        embeddings_dir.mkdir(parents=True, exist_ok=True)
        
        # Separar chunks y embeddings
        chunks_data = [{
            "content": chunk["content"],
            "metadata": chunk["metadata"]
        } for chunk in chunks]
        
        embeddings_data = [{
            "id": i,
            "embedding": chunk["embedding"]
        } for i, chunk in enumerate(chunks)]
        
        # Guardar chunks en formato JSON
        chunks_file = chunks_dir / f"chunks_{timestamp}.json"
        with open(chunks_file, 'w', encoding='utf-8') as f:
            json.dump(chunks_data, f, ensure_ascii=False, indent=2)
        
        # Guardar embeddings en formato pickle
        embeddings_file = embeddings_dir / f"embeddings_{timestamp}.pkl"
        with open(embeddings_file, 'wb') as f:
            pickle.dump(embeddings_data, f)
        
        return chunks_file, embeddings_file

# Uso del procesador
def process_text(splits):
    processor = TextProcessor()
    
    # 1. Realizar segmentación semántica
    print("Realizando segmentación semántica...")
    chunks = processor.semantic_chunk(splits)
    print(f"Se generaron {len(chunks)} chunks")
    
    # 2. Generar embeddings
    print("\nGenerando embeddings...")
    processed_chunks = processor.generate_embeddings(chunks)
    
    # 3. Guardar resultados
    print("\nGuardando resultados...")
    chunks_file, embeddings_file = processor.save_processed_data(processed_chunks)
    
    print(f"\nProceso completado:")
    print(f"Chunks guardados en: {chunks_file}")
    print(f"Embeddings guardados en: {embeddings_file}")
    
    return chunks_file, embeddings_file

# Procesar los splits
chunks_file, embeddings_file = process_text(splits)
```

```python
from qdrant_client import QdrantClient
from qdrant_client.models import PointStruct, Distance, VectorParams
import numpy as np
import json

# Configuración de Qdrant
qdrant_client = QdrantClient(":memory:")  # Cambia ":memory:" por la dirección de tu servidor Qdrant si es remoto

# Nombre de la colección
collection_name = "pdf_chunks"

# Parámetros de los vectores
embedding_dim = 384  # Dimensión de los embeddings generados por el modelo
qdrant_client.recreate_collection(
    collection_name=collection_name,
    vectors_config=VectorParams(size=embedding_dim, distance=Distance.COSINE),
)

# Cargar embeddings y chunks guardados
embeddings = np.load("embeddings/embeddings.npy")
with open("data/chunks/chunks.json", "r", encoding="utf-8") as f:
    chunks = json.load(f)

# Indexar embeddings en Qdrant
points = [
    PointStruct(
        id=i,
        vector=embedding.tolist(),
        payload={"text": chunk["text"], "doc_id": chunk["doc_id"]},
    )
    for i, (embedding, chunk) in enumerate(zip(embeddings, chunks))
]
qdrant_client.upsert(collection_name=collection_name, points=points)

print("Embeddings indexados en Qdrant.")
```

```python
from transformers import AutoTokenizer, AutoModel

# Función para generar embeddings de consultas
def generate_query_embedding(query, model, tokenizer):
    inputs = tokenizer(query, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
        query_embedding = outputs.last_hidden_state.mean(dim=1)
    return query_embedding[0].numpy()

# Función para realizar consultas en Qdrant
def search_qdrant(query, top_k=5):
    query_embedding = generate_query_embedding(query, model, tokenizer)
    results = qdrant_client.search(
        collection_name=collection_name,
        query_vector=query_embedding.tolist(),
        limit=top_k,
    )
    return results

# Ejemplo de consulta
query = "¿Qué dice el documento sobre regulación?"
results = search_qdrant(query)

# Mostrar resultados
print("Resultados de la consulta:")
for result in results:
    print(f"Score: {result.score:.4f}, Text: {result.payload['text']}")
```
